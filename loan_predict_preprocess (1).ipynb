```python
xxf=xf.values.reshape(-1,5)
xxt=xt.values.reshape(-1,5)
yt=ytest.values.reshape(-1,1)
ytar=ytrain.values.reshape(-1,1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
mo=KNeighborsClassifier()
```

```python
ml=mo.fit(xxf,ytar)
```

```python
yp=ml.predict(xxt)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(yt,yp)*100
```
